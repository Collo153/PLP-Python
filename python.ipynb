
# Python Data Analysis Assignment

This notebook contains solutions for:
- Data loading and exploration
- Basic data analysis
- Data visualization


```

import pandas as pd

# Load dataset (Iris dataset from sklearn for demonstration)
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
df = iris.frame

# Display first few rows
print("First 5 rows:")
display(df.head())

# Check structure
print("\nData info:")
print(df.info())

# Check missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```

# Statistical summary
print("Statistical summary:")
display(df.describe())

# Group by target (species) and compute mean
print("\nMean values per species:")
display(df.groupby("target").mean())
```

```

import matplotlib.pyplot as plt
import seaborn as sns

# Line chart - trend over index (not time-based but for demo)
plt.figure(figsize=(8,4))
plt.plot(df.index, df['sepal length (cm)'])
plt.title("Line Chart: Sepal Length over Index")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()

# Bar chart - average petal length per species
plt.figure(figsize=(6,4))
df.groupby("target")['petal length (cm)'].mean().plot(kind="bar")
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()

# Histogram - sepal width distribution
plt.figure(figsize=(6,4))
plt.hist(df['sepal width (cm)'], bins=20, edgecolor='black')
plt.title("Histogram: Sepal Width Distribution")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()

# Scatter plot - sepal length vs petal length
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['sepal length (cm)'], y=df['petal length (cm)'], hue=df['target'])
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```
